```python
from transformer_lens import HookedTransformer
from mechinterp import mini_tools  # TODO modify when adding `mini_tools` to Interpreter class

message = "What does `pip install mechinterp` do?"
model_name = "google/gemma-2-2b-it"

model = HookedTransformer.from_pretrained_no_processing(model_name)
```

```python
# generate model's response to inspect model activations, while including it
full_chat_toks, full_chat_str, response_str = mini_tools.generate(model, message, max_new_tokens=28)

print(full_chat_str, "[...]")
```

```python
# inspects a specific direction along this generation
direction = torch.randn(2304)


fig, sims = mini_tools.cosine_with_direction(model, full_chat_toks, direction)
```
